### File not explicity closed is not a good practice. Some time, we will lose data as it will still be in buffer and not written down to the file. May run out of sys resources or other corruption issues. Always a good practice to close the file 

###  Good alernative is to use with block. This is python inbuilt tool to ensure file is closed always.

```python
with open("test.txt",mode = "at",encoding="utf-8") as fappend:
    fappend.write("new entry")
    x = 1 
    y = 0 
    fappend.write("division of x by y ",x/y)
```

```python
# How to check if the file is open or closed.

if fappend.closed:
    print("The file is closed.")
else:
    print("The file is open in " , fappend.mode , " mode")
```

#### Binary File 

```python
# Open a binary file for reading and writing
# No encoding as it doesnt make sense for binary files

with open('file.bin', 'wb') as f:
    # Read and write to the file as you would with a regular file
    f.write(b'Hello, world!')
    
```

```python
# b represents binary data. Although data is stored in the file in binary format.

with open('file.bin', 'rb') as f:
    f.seek(0)
    print(f.read())
```

#### n Python, the b prefix at the beginning of a string literal indicates that the string is a "byte string". It is used to specify that the string should be stored as a series of bytes, rather than as a series of Unicode characters.
