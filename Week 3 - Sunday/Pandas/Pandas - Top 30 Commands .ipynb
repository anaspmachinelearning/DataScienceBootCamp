## missing values

```python
titanic.info()
```

```python
# A quicker way to get the null value count for each column

#titanic.isna()
titanic.isna().sum()
```

```python
# various ways to deal with missing values 

# Drop the column entirely or set a threshold

titanic.dropna(axis='columns').head()
```

```python
# Age column survived!

titanic.dropna(thresh=len(titanic)*0.8, axis='columns').head()

```

```python
### Row drop - NaN values


titanic.dropna( axis='index').info()
```

```python
# Fill with some values 

titanic["Cabin"].fillna("Default Cabin" , inplace = True)
```

```python
titanic.info()
```

```python
# Get the value count for how many were replaced.
titanic.Cabin.value_counts()
```

## Sorting

### Aggregation 

```python
titanic.head()
```

```python
# Those who survived what was their mean Fare
#titanic[titanic.Survived == 1]
titanic[titanic.Survived == 1].Fare.mean()
```

```python
# Those who didn't survive what was their mean Fare

titanic[titanic.Survived == 0].Fare.mean()
```

```python
titanic.groupby('Survived').Fare.agg(['sum', 'count','mean']).head()
```
